### Conexão com o mongo

```python
mongo_client = MongoClient(_connection_url)
db = mongo_client[_db_cartola]
```

### Função que retorna uma lista de atletas pelo número da rodada

```python
def get_atletas_by_rodada(collection,rodada):
    document = db[_collection].find({'id_rodada':rodada})
    atletas = dict([i for i in document][0])['atletas']
    return atletas
```

## Lendo dados da rodada com panda

```python
data = pd.DataFrame(get_atletas_by_rodada(_collection,_id_rodada))
data
```
