### Função que retorna uma lista de atletas pelo número da rodada

```python
def get_atletas_by_rodada(collection,rodada):
    document = db[_collection].find({'id_rodada':rodada})
    atletas = dict([i for i in document][0])['atletas']
    return atletas
```

<h3> Função que retorna uma lista com todos os atletas e escouts de todas as rodadas </h3>

```python
def get_atletas():
    document = db[_collection].find()    
    atletas = list()
    #loop na collection (cada item corresponde a uma rodada)
    for j in range(0,document.count()):
        #lista de atletas
        data      = document[j]['atletas'] 
        #preenche id da rodada
        id_rodada = document[j]['id_rodada'] 
        #adiciona o id da rodada no dicionário de atletas
        for l in data:
            l['rodada'] = id_rodada        
        #adiciona os itens da nova lista à lista existente
        atletas.extend(data)        
    return atletas
```

<h3> Pre Processing and Data Cleaning </h3>

```python
#adiciona colunas no DataFrame para representação abreviada dos scouts
def process_scout(row):
    scout = row['scout']
    had_scout=0
    for i in scout.items():
        scout = i[1]
        row[i[0]] = scout
        
        if ( not(numpy.isnan( scout ) ) & ( scout != 0 ) ):
            had_scout = 1
    
    row['had_scout'] = had_scout
    return row
```

```python
#Dict com os nomes das posições
posicao = {1:'Goleiro',2:'Lateral',3:'Zagueiro',4:'Meia',5:'Atacante',6:'Técnico'}

#função para plotar a média de pontuação de cada uma das 6 posições dos 20 clubes
def plot_mean_ratings_clubs_bar(sframe,graph_type):
    #cria o multi-gráfico (2 linhas, 3 colunas)
    fig,axes = plt.subplots(2,3,sharey=True,figsize=(15,10))
    #configura o espaço entre as duas linhas
    fig.subplots_adjust(hspace=.5)
    #loop para preencher cada gráfico
    for n in range(len(axes.flat)):
        #obtém o eixo (gráfico do subplot)
        ax = axes.flat[n]
        #filtra os jogadores por posição
        data = sframe[sframe['pos']==n+1]
        #agrupa as pontuações das posições por clube e ordena o resultado
        grouped_mean = data['pontos'].groupby(data['clube']).mean().sort_values(axis=0,ascending=False)
        grouped_dev = data['pontos'].groupby(data['clube']).std().sort_values(axis=0,ascending=False)
        if graph_type == 'bar':
            #plota o gráfico de barras
            ax.bar(left=range(1,grouped_mean.count()+1), height=grouped_mean.values )
            #adiciona os pontos no eixo x (20 pontos, referente aos 20 clubes da série A)
            ax.set_xticks([i for i in range(1,21)])
            #adiciona os nomes dos clubes
            ax.set_xticklabels(grouped_mean.index)
        elif graph_type == 'std':
            ax.plot(range(1,grouped_dev.count()+1),grouped_dev.values,color='r')
            #adiciona os pontos no eixo x (20 pontos, referente aos 20 clubes da série A)
            ax.set_xticks([i for i in range(1,21)])
            #adiciona os nomes dos clubes
            ax.set_xticklabels(grouped_dev.index)            
        #Adiciona o título 
        ax.set_title(posicao[n+1])
        #Adiciona o label do eixo y
        ax.set_ylabel('Pontuação Média')
        #Adiciona o label do eixo x
        ax.set_xlabel('Clubes')
        
        #rotaciona os nomes para melhorar a visualização
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
```

```python
from sklearn.cross_validation import KFold
from sklearn.linear_model import LinearRegression
#from sklearn.linear_model import Ridge
#from sklearn.cross_validation import train_test_split

def linear_regression_cross_validated(dataset,feature_names,target):
    linear_data = dataset[feature_names].copy()
    target = dataset[target]
    k_fold = KFold(n=len(linear_data),n_folds=5)
      
    min_error = 1e15    
    model=None
    
    for train_indices, test_indices in k_fold:
        train_set = numpy.array( linear_data.iloc[train_indices] )
        test_set  = numpy.array( linear_data.iloc[test_indices] )
        
        train_output    = numpy.array( target.iloc[train_indices])
        test_output     = numpy.array( target.iloc[test_indices])
        
        lr = LinearRegression()
        lr.fit(train_set,train_output)
        
        predicted_output = lr.predict(test_set)
        error = lr.score(test_set,test_output)
        #error = numpy.sum((predicted_output-test_output)**2)
        
        #er2 = r2_score(test_output,predicted_output)
            
        if error < min_error:
            model = lr
            min_error = error
            
    return model
```
