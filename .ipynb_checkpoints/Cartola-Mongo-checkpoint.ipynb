### configurações e definições

```python
#URL do mongodb
_connection_url = 'mongodb://localhost:27017'
#nome do banco mongodb
_db_cartola='cartola-fc'
#nome da collection mongodb
_collection='mercado'
#id da rodada
_id_rodada=17
```

### Conexão com o mongo

```python
mongo_client = MongoClient(_connection_url)
db = mongo_client[_db_cartola]
```

### Função que retorna uma lista de atletas pelo número da rodada

```python
def get_atletas_by_rodada(collection,rodada):
    document = db[_collection].find({'id_rodada':rodada})
    atletas = dict([i for i in document][0])['atletas']
    return atletas
```

<h3> Função que retorna uma lista com todos os atletas e escouts de todas as rodadas </h3>

```python
def get_atletas():
    document = db[_collection].find()    
    atletas = list()
    #loop na collection (cada item corresponde a uma rodada)
    for j in range(0,document.count()):
        #lista de atletas
        data      = document[j]['atletas'] 
        #preenche id da rodada
        id_rodada = document[j]['id_rodada'] 
        #adiciona o id da rodada no dicionário de atletas
        for l in data:
            l['rodada'] = id_rodada        
        #adiciona os itens da nova lista à lista existente
        atletas.extend(data)        
    return atletas
```

<h2> Leitura dos dados dos atletas com pandas.DataFrame </h2>

```python
frame = pd.DataFrame(get_atletas())
#frame[frame['clube']=='Corinthians']
```

<h3> Pre Processing and Data Cleaning </h3>

```python
# Função que retorna o código da posição de um atleta
def process_position(row):
    pos = row['posicao']
    return pos[0]

# Função que retorna o código do status de um atleta
def process_status(row):
    st = row['status']
    return st[0]

#adiciona colunas no dataframe para representação completa
def process_total_rating(row):
    scout = row['scout']
    for i in scout.items():
        rating = scouts_type[i[0]]
        row[rating[0]] = rating[1]*i[1]
    return row

#adiciona colunas no DataFrame para representação abreviada dos scouts
def process_scout(row):
    scout = row['scout']
    for i in scout.items():
        #rating = scouts_type[i[0]]
        row[i[0]] = i[1]
    return row
```

<h2> Aplica as funções de processamento e limpeza dos dados </h2>

```python
#detalhes de todos os scouts possíveis
scouts_details = {
    'RB': ('Roubadas de Bola',+1.7),
    'FC': ('Falta Cometida',-0.5),
    'GC': ('Gol Contra',-6.),
    'CA': ('Cartão Amarelo',-2.),
    'CV': ('Cartão Vermelho',-5.),
    'SG': ('Jogos sem sofrer Gols',+5.),
    'DD': ('Defesa Difícil',+3.),
    'DP': ('Defesa de Penalti',+7.),
    'GS': ('Gol Sofrido',-2.),
    'FS': ('Falta Sofrida',+0.5),
    'PE': ('Passe Errado',-0.3),
    'A' : ('Assistencia',+5.),
    'FT': ('Finalizacao na trave',+3.5),
    'FD': ('Finalizacao Defendida',+1.),
    'FF': ('Finalizacao Fora',+0.7),
    'G' : ('Gol',+8.),
    'I' : ('Impedimento',-0.5),
    'PP': ('Penalti Perdido',-3.5)
}

#scouts de defesa (jogadores de linha)
defense_scouts = {scouts_details['RB'],scouts_details['FC'],scouts_details['GC'],scouts_details['CA'],
                  scouts_details['CV'],scouts_details['SG'],scouts_details['GS'],scouts_details['FS'],scouts_details['PE']}


data_labels = ['apelido','clube','pos','st','rodada','pontos']
```

```python
#remove as duplicidades
frame = frame.drop_duplicates(subset=['rodada','pontos','apelido','clube','id'])
#cria uma coluna com um valor numérico para a posicao
frame['pos'] = frame.apply(func=process_position,axis=1)
#cria uma coluna com um valor numérico do status
frame['st']  = frame.apply(func=process_status,axis=1)
#cria colunas com os scouts dos atletas
frame = frame.apply(process_scout,axis=1)
#preenche os valores não numéricos
frame.fillna(value=0.,inplace=True)
#preserva apenas as colunas mais importantes do dataset
column_names = data_labels
column_names.extend(scouts_details)
frame = frame[column_names]
```

```python
# Separa os jogadores que não tiveram nenhum scout registrado até este momento no campeonato.
inativos = frame[frame['pontos']==0.00].copy()
# Remove os jogadores que não pontuaram
frame = frame[(frame['pontos']!=0.0)]
```

```python
#Dict com os nomes das posições
posicao = {1:'Goleiro',2:'Lateral',3:'Zagueiro',4:'Meia',5:'Atacante',6:'Técnico'}

#função para 
def plot_mean_ratings_clubs_bar(sframe,graph_type):
    #cria o multi-gráfico (2 linhas, 3 colunas)
    fig,axes = plt.subplots(2,3,sharey=True,figsize=(15,10))
    #configura o espaço entre as duas linhas
    fig.subplots_adjust(hspace=.5)
    #loop para preencher cada gráfico
    for n in range(len(axes.flat)):
        #obtém o eixo (gráfico do subplot)
        ax = axes.flat[n]
        #filtra os jogadores por posição
        data = sframe[sframe['pos']==n+1]
        #agrupa as pontuações das posições por clube e ordena o resultado
        grouped_mean = data['pontos'].groupby(data['clube']).mean().sort_values(axis=0,ascending=False)
        grouped_dev = data['pontos'].groupby(data['clube']).std().sort_values(axis=0,ascending=False)
        if graph_type == 'bar':
            #plota o gráfico de barras
            ax.bar(left=range(1,grouped_mean.count()+1), height=grouped_mean.values )
            #adiciona os pontos no eixo x (20 pontos, referente aos 20 clubes da série A)
            ax.set_xticks([i for i in range(1,21)])
            #adiciona os nomes dos clubes
            ax.set_xticklabels(grouped_mean.index)
        elif graph_type == 'std':
            ax.plot(range(1,grouped_dev.count()+1),grouped_dev.values,color='r')
            #adiciona os pontos no eixo x (20 pontos, referente aos 20 clubes da série A)
            ax.set_xticks([i for i in range(1,21)])
            #adiciona os nomes dos clubes
            ax.set_xticklabels(grouped_dev.index)            
        #Adiciona o título 
        ax.set_title(posicao[n+1])
        #Adiciona o label do eixo y
        ax.set_ylabel('Pontuação Média')
        #Adiciona o label do eixo x
        ax.set_xlabel('Clubes')
        
        #rotaciona os nomes para melhorar a visualização
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=90)
```

```python
plot_mean_ratings_clubs_bar(frame,'bar')
plot_mean_ratings_clubs_bar(frame,'std')
```

```python
from sklearn.cross_validation import KFold
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.cross_validation import train_test_split

def linear_regression(dataset,feature_names,target):
    linear_data = dataset[feature_names].copy()
    target = dataset[target]
    k_fold = KFold(n=len(linear_data),n_folds=5)
      
    min_error = 1e15    
    model=None
    
    for train_indices, test_indices in k_fold:
        train_set = numpy.array( linear_data.iloc[train_indices] )
        test_set  = numpy.array( linear_data.iloc[test_indices] )
        
        train_output    = numpy.array( target.iloc[train_indices])
        test_output     = numpy.array( target.iloc[test_indices])
        
        lr = LinearRegression()
        lr.fit(train_set,train_output)
        
        predicted_output = lr.predict(test_set)        
        error = numpy.sum((predicted_output-test_output)**2)
        
        #er2 = r2_score(test_output,predicted_output)
            
        if error < min_error:
            model = lr
            min_error = error
            
    return model
```

```python
train , test = train_test_split(frame, test_size=0.15, random_state=10)
model1 = linear_regression(train,list(scouts_details.keys()),'pontos')

print(model1.coef_)
```

```python
test_set   =  test[list(scouts_details.keys())]
test_array  = numpy.array(test_set)

predict = model1.predict(test_array)
```
