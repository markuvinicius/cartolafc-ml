```python
import graphlab
import pandas as pd
import numpy
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#pd.read_csv?
atletas = graphlab.SFrame('atletas2.csv')
```

```python
atletas.head()
```

```python
features = ['R%d' %i for i in range(1,12)]
ratings = atletas[features].copy()
```

```python
ratings['target'] = atletas['R12']
ratings.head()
```

```python
def polynomial_sframe(feature, degree):

    # initialize the SFrame:
    poly_sframe = graphlab.SFrame()
    # and set poly_sframe['power_1'] equal to the passed feature
    poly_sframe['power_1'] = feature
    # first check if degree > 1
    if degree > 1:
        # then loop over the remaining degrees:
        # range usually starts at 0 and stops at the endpoint-1. We want it to start at 2 and stop at degree
        for power in range(2, degree+1): 
            # first we'll give the column a name:
            name = 'power_' + str(power)
            # then assign poly_sframe[name] to the appropriate power of feature
            poly_sframe[name] = feature.apply(lambda x:x**power)
            
    return poly_sframe    
```

```python
def polynomial_regression(data,l2_penalty,degree):
    
    features = ['power_%d'%i for i in range(1,degree+1)]
    
    model = graphlab.linear_regression.create(dataset=data,
                                              target='target',
                                              features=features,
                                              verbose=False,
                                              validation_set=None,
                                              l2_penalty=l2_penalty
                                             )
    return model
```

```python
def print_coefficients(model):    
    # Get the degree of the polynomial
    deg = len(model.coefficients['value'])-1

    # Get learned parameters as a list
    w = list(model.coefficients['value'])

    # Numpy has a nifty function to print out polynomials in a pretty way
    # (We'll use it, but it needs the parameters in the reverse order)
    print 'Learned polynomial for degree ' + str(deg) + ':'
    w.reverse()
    print numpy.poly1d(w)
```

```python
def plot_model(data,model):
    
    y_predicted = model.predict(data)
    plt.Figure()
    plt.plot(data['power_1'],data['target'],'.',c='b')
    plt.plot(data['power_1'],y_predicted,'-',c='r')
```

```python
def row_mean(row):
    features = ['R%d' %i for i in range(1,12)]
    s=0
    for f in features:
       s+=row[f]
    return (s/11)
```

```python
ratings['média'] = ratings.apply(row_mean)
ratings
```

```python
ratings = ratings.sort(['média'],ascending=False).dropna()


degree = 9
data = polynomial_sframe(ratings['média'],degree)
data['target']=ratings['target']
print data

l2_small_penalty = 0
model = polynomial_regression(data,l2_small_penalty,degree)
print_coefficients(model)
plot_model(data,model)
```
